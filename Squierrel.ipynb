```python
import pandas as pd
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
squierrel = pd.read_csv('2018_Central_Park_Squirrel_Census_-_Squirrel_Data.csv')
```

```python
squierrel['Date'] = pd.to_datetime(squierrel['Date'], format='%m%d%Y')
```

```python
squierrel
```

```python
squierrel.dtypes
```

```python
grouped = squierrel.groupby(pd.Grouper(key='Date', freq='D')).nunique()
```

```python
grouped.dtypes
```

```python
grouped.plot(y='Unique Squirrel ID')
```

```python
groupedCount= squierrel.groupby(pd.Grouper(key='Date', freq='D')).count()
```

```python
groupedCount.plot(y='Unique Squirrel ID')
```
